Information Theroy
---------------------------

3.1) $$Gain(X,X) := I(X,Y) := H(X) - H(X|X) = H(X)$$

In information gain, subtraction term is for the reduction of uncertainty about X. But, X given X is not giving any conditional information, which means $H(X|X) = 0$.
Therefore, the expected information gain between X and itself is just $H(X)$.

3.2) 

1. Each events happening separely

$$P((X|E_2)|(E_1|E_2)) = \frac{P((X|E_2), (E_1|E_2))}{P(E_1|E_2)}=\frac{P(X,E_1|E_2)}{P(E_1|E_2)} = \frac{P(X,E_1,E_2)}{P(E_1|E_2)P(E_2)} = \frac{P(X,E_1,E_2)}{P(E_1)P(E_2)}$$



By Bayes' theorem which is $$P(A|B) = \frac{P(B|A)P(A)}{P(B)}$$, the equation, $$\frac{P(X,E_1|E_2)}{P(E_1|E_2)} = \frac{P(X,E_1,E_2)}{P(E_1|E_2)P(E_2)}$$ can be proved. The reason why we got final term from third term is because $E_1$ and $E_2$ are independent event. Therefore, $P(E_1|E_2) = P(E_1)$.

2. events happening altogether

$$P(X|E_1,E_2) = \frac{P(X,E_1,E_2)}{P(E_1,E_2)} = \frac{P(X,E_1,E_2)}{P(E_1)P(E_2)}$$

Again, event $E_1$ and $E_2$ are independent events, so from this we can get $$P(E_1,E_2) = P(E_1)P(E_2)$$.

Because the information gain for first two events ($E_1, E_2$) is same for each case, for the rest of all events it's proved that the total information gain of each of these events happending separately is the same as the information gain of all these events happening together. 

3.3) $$I(X,Y) = H(Y) - H(Y|X)$$
$$I(X,Y) = H(Y) - H(Y|X) \geq 0$$
$$H(Y) \geq H(Y|X)$$

For the right hand side $H(Y|X)$, in the maximum case Y and X are independent. 
If these two are independent, $H(Y|X) = H(Y)$.It will still meet the equation with $H(Y) \geq H(Y)$. Therefore, $I(X,Y) \geq 0$ is true.

3.4) $$Z = X + Y$$
$$P(Z|X) = P(Z-X|X) = P(Y|X)$$
$$P(Z|X) = -\sum\limits_{x}\sum\limits_{z}P(z|x)\log_{10} P(z|x)$$
Because P(z|x) = P(y|x),

$$ = -\sum\limits_{x}P(x){\sum\limits_{y}P(y|x)\log_{10}P(y|x)} = H(Y|X)$$

$$G.I = 2p(1-p) = 2({1 \over 2})({1 \over 2})$$

```python
# Decision Tree Implementations from sklearn CART
from IPython.display import SVG
from graphviz import Source #graphviz can be used to visualize the tree built

import pandas as pd #data processing, CSV file I/O

from sklearn import tree
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt

import numpy as np #linear algebra

#Load Dataset
data = pd.read_csv("Titanic.csv")

#Inspect Values of each feature for processing
data.info()
#With the information of data checking, it's realized that Cabin has many missing data. Decided to get Has_Cabin instaed of just Cabin data.

'''Encode non-numeric features'''
'''Convert the non-numerical columns Sex, Cabin, and Embarked into numerical values'''

# Encode Sex features to numerical values
s1 = pd.Series({'male':0, 'female':1})
data['Sex']=data['Sex'].map(s1)

# Encode Embarked features to numerical values
s2= pd.Series({'S':0, 'C':1, 'Q':2})
data['Embarked']=data['Embarked'].map(s2)

# Get Has_Cabin feature instead of just Cabin
data['Has_Cabin'] = data['Cabin'].apply(lambda x: 0 if type(x)==float else 1) # get 1 if the passenger had cabin else get 0


'''Remove feature with no useful information and drop NaN'''
data.pop('Cabin') # remove Cabin feature
data.dropna(inplace=True) # drop the rows that have missing data

# relabeling data in order after dropping useless data
data = data.reset_index()
data = data.drop(['index'],axis=1)


'''Seperate features (X) and targets (y)'''

# set up the train data with columns that are in use
train_pre = data[['Pclass','Sex','Age','SibSp','Parch','Fare','Has_Cabin','Embarked']]

# split data with train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_pre, data[['Survived']],random_state=13)

# Get the decision tree from DecisionTreeClassifier
tree_clf = tree.DecisionTreeClassifier(max_depth=3, random_state = 13)
tree_clf.fit(X_train, y_train)

# Check the accuracy
print('Score: {}'.format(tree_clf.score(X_train, y_train)))

# visualize the decision tree
from sklearn.tree import export_graphviz

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(tree_clf,
                  feature_names=['Pclass','Sex','Age','Sibsp','Parch','Fare','Has_Cabin','Embarked'],
                  class_names=['Unsurvived','Survived'],
                  filled=True)

# save the image of decision tree as png type
fig.savefig("Titanic_decision_tree.png")
```
